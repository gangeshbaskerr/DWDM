```python
def gen_basket(no_of_recs):

    t_apple = rd.choices(m_fruits.apple,k=no_of_recs)
    t_kiwi = rd.choices(m_fruits.kiwi,k=no_of_recs)
    t_bread = rd.choices(m_dairyproducts.bread,k=no_of_recs)
    t_cheese = rd.choices(m_dairyproducts.cheese,k=no_of_recs)
    t_dairymilk=rd.choices(m_chocolate.dairymilk,k=no_of_recs)
    t_kindercreamy=rd.choices(m_chocolate.kindercreamy,k=no_of_recs)
    t_teddy=rd.choices(m_toys.teddy,k=no_of_recs)
    t_talkingtom=rd.choices(m_toys.talkingtom,k=no_of_recs)
    t_tomato=rd.choices(m_vegetables.tomato,k=no_of_recs)
    t_onion=rd.choices(m_vegetables.onion,k=no_of_recs)
    cube = pd.DataFrame({'apple':t_apple,
                      'kiwi':t_kiwi,
                         'bread':t_bread,
                         'cheese':t_cheese,
                         'dairymilk':t_dairymilk,
                         'kindercreamy':t_kindercreamy,
                         'teddy':t_teddy,
                         'talkingtom':t_talkingtom,
                         'tomato':t_tomato,
                         'onion':t_onion,
                         }
                      )
    return(cube)
```

```python
test =gen_basket(50)
```

```python
def amount(cols):
    value = cols[0]
    a = cols[1]
    b =cols[2]
    c =cols[3]
    d =cols[4]
    e =cols[5]
    f =cols[6]
    g =cols[7]
    h =cols[8]
    i =cols[9]
    j =cols[10]
    while value == 1:
     amountt=a*25+b*30+c*5+d*20+e*10+f*30+g*150+h*200+i*6+j*4
     return amountt
```

```python
test["amount"] = test[['value','apple','kiwi','bread','cheese','dairymilk','kindercreamy','teddy','talkingtom','tomato','onion']].apply(amount,axis=1)
```
