```python
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
n_samples = 100
X = np.random.randn(n_samples, 2)
n_clusters = 4
```

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(121)
for cluster_id in range(n_clusters):
    plt.scatter(X[kmeans_labels == cluster_id, 0], X[kmeans_labels == cluster_id, 1], label=f'Cluster {cluster_id + 1}')
plt.title("K-Means Clustering")
plt.legend()

plt.subplot(122)
unique_labels = np.unique(dbscan_labels)
for label in unique_labels:
    if label == -1:
        plt.scatter(X[dbscan_labels == label, 0], X[dbscan_labels == label, 1], label='Outliers', marker='x')
    else:
        plt.scatter(X[dbscan_labels == label, 0], X[dbscan_labels == label, 1,], label=f'Cluster {label}')
plt.title("DBSCAN Clustering")
plt.legend()

plt.show()
```
